```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Loading and cleaning the data
planets = sns.load_dataset('planets')
planets_clean = planets.dropna()


# 1. Line Chart: Total Discoveries by Year
yearly_discoveries = planets['year'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
plt.plot(yearly_discoveries.index, yearly_discoveries.values, marker='o')
plt.title('Line Chart: Planet Discoveries per Year')
plt.xlabel('Year')
plt.ylabel('Number of Discoveries')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 2. Bar Chart: Avg Mass of Planets by Discovery Method
avg_mass_by_method = planets_clean.groupby('method')['mass'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
avg_mass_by_method.plot(kind='bar', color='skyblue')
plt.title('Bar Chart: Average Planet Mass by Discovery Method')
plt.ylabel('Average Mass (Jupiter Mass)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# 3. Histogram:
numeric_cols = ['number', 'orbital_period', 'mass', 'distance']

for col in numeric_cols:
    plt.figure(figsize=(8, 5))
    plt.hist(planets_clean[col], bins=40, color='cornflowerblue')
    plt.title(f'Histogram: {col.capitalize()} Distribution')
    plt.xlabel(col.capitalize())
    plt.ylabel('Frequency')
    if col != 'number':
        plt.xscale('log')
    plt.tight_layout()
    plt.show()
```

```python
# 4. Pie Chart: Planet Counts by Number of Planets in System
system_counts = planets_clean['number'].value_counts().head(5)
others = planets_clean['number'].value_counts().iloc[5:].sum()
pie_data = pd.concat([system_counts, pd.Series({'Other': others})])

plt.figure(figsize=(7, 7))
plt.pie(pie_data, labels=pie_data.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart: Number of Planets in System')
plt.tight_layout()
plt.show()
```

```python
# 5. Box & Whiskers:
for col in ['number', 'mass', 'orbital_period', 'distance']:
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=planets_clean, x='method', y=col, palette='Set2')
    plt.xticks(rotation=90)
    if col != 'number':
        plt.yscale('log')
    plt.title(f'Box & Whiskers: {col.capitalize()} by Method')
    plt.tight_layout()
    plt.show()
```

```python
# 6. Scatter Plots:

# Mass vs Orbital Period
plt.figure(figsize=(10, 6))
sns.scatterplot(data=planets_clean, x='orbital_period', y='mass', hue='method', alpha=0.7)
plt.xscale('log')
plt.yscale('log')
plt.title('Scatter Plot: Mass vs Orbital Period')
plt.tight_layout()
plt.show()

# Mass vs Distance
plt.figure(figsize=(10, 6))
sns.scatterplot(data=planets_clean, x='distance', y='mass', hue='method', alpha=0.7)
plt.xscale('log')
plt.yscale('log')
plt.title('Scatter Plot: Mass vs Distance')
plt.tight_layout()
plt.show()

# Mass vs Number of Planets in System
plt.figure(figsize=(10, 6))
sns.scatterplot(data=planets_clean, x='number', y='mass', hue='method', alpha=0.7)
plt.yscale('log')
plt.title('Scatter Plot: Mass vs Number of Planets in System')
plt.tight_layout()
plt.show()
```

```python
# 7. Count Plot: Discoveries per Year

plt.figure(figsize=(12, 5))
sns.countplot(data=planets, x='year', palette='coolwarm')
plt.xticks(rotation=45)
plt.title('Count Plot: Discoveries by Year')
plt.tight_layout()
plt.show()
```

```python
# 8. Heatmap: Missing Values

plt.figure(figsize=(10, 4))
sns.heatmap(planets.isnull(), cmap='Blues', cbar=False)
plt.title('Heatmap: Missing Data in Planets Dataset')
plt.tight_layout()
plt.show()
```
